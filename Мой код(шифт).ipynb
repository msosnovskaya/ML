```python
data=pd.read_csv("train.csv")
```

```python
X = data.drop(columns=['churn'])
y = data['churn']
```

```python
X=X.drop(columns=['numbcars', 'dwllsize','HHstatin','lor'])
```

```python
float_ = X.rev_Mean.dtype #заполняем пропуски 
int_ = X.avg3rev.dtype
obj_ = X.ethnic.dtype
for x in X.columns:
    if (X[x].dtype==float_)| (X[x].dtype==int_):
        X[x] = X[x].fillna(X[x].median())
    if (X[x].dtype==obj_):
        X[x] = X[x].fillna('UNKNOWN')
```

```python
cat_col = list(data.select_dtypes(include='object'))
```

```python
import catboost
from catboost import CatBoostClassifier
from sklearn.model_selection import KFold, cross_val_score
model = CatBoostClassifier(loss_function='Logloss',eval_metric="AUC")
kf = KFold(n_splits=5, random_state=42) # cоздаем генератор разбиений
a=cross_val_score(model, X, y, cv=kf,scoring='roc_auc')
a.mean()
```
